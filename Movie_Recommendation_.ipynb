```python
#importing the libraries
import pandas as pd
import numpy as np
```

```python
#importing the dataset
movies = pd.read_csv('tmdb_5000_movies.csv')
credits = pd.read_csv('tmdb_5000_credits.csv')
```

```python
movies.head(1)
```

```python
credits.head(1)['cast'].values
```

```python
#Merging the two datsets together
movies = movies.merge(credits,on='title')
```

```python
movies.shape
```

```python
#extracting the relevant columns
#genres
#id
#keywords
#title
#overview
#cast
#crew
movies = movies[['id','title','genres','overview','keywords','cast','crew']]
```

```python
#checking the Null values
movies.isnull().sum()
```

```python
# Removing the empty values
movies.dropna(inplace=True)
```

```python
#checking for duplicates
movies.duplicated().sum()
```

```python
movies['genres'][0].split(',')
```

```python
import ast
```

```python
ast.literal_eval('[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]')
```

```python
# creating the function to get the genre details from genre column
def convert(obj):
  L = []
  for i in ast.literal_eval(obj):
    L.append(i['name'])
  return L
```

```python
# applying the function to the genre column
movies['genres']=movies['genres'].apply(convert)
```

```python
movies.head()
```

```python
# applying the function to remaining columns i.e keywords
movies['keywords'] = movies['keywords'].apply(convert)
```

```python
movies.head()
```

```python
#selecting the first three cast members from the Cast column
def convert_2(obj):
  L = []
  counter = 0
  for i in ast.literal_eval(obj):
    L.append(i['name'])
    counter += 1
    if counter == 3:
      break
  return L
```

```python
movies['cast']=movies['cast'].apply(convert_2)
```

```python
#selecting the director name from the crew column
def fetch_director(obj):
  L = []
  for i in ast.literal_eval(obj):
    if i['job'] == 'Director':
      L.append(i['name'])
  return L
```

```python
#applying the function to the crew column
movies['crew']=movies['crew'].apply(fetch_director)
```

```python
movies.head()
```

```python
# Creating a function to remove the space between words
def space_remover(obj):
  l = []
  for i in obj:
    l.append(i.replace(" ",""))
  return l
```

```python
# applying the function to all the columns
movies['genres']=movies['genres'].apply(space_remover)
movies['keywords']=movies['keywords'].apply(space_remover)
movies['cast']=movies['cast'].apply(space_remover)
movies['crew']=movies['crew'].apply(space_remover)
```

```python
# Converting the overview column to a list
movies['overview']=movies['overview'].apply(lambda x:x.split())
```

```python
movies.head()
```

```python
#concatinating all the columns to make a new column named Tags
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
```

```python
#creating new dataframe
new_df = movies[['id','title','tags']]
```

```python
new_df.head()
```

```python
#converting the tags list into string
new_df['tags'] = new_df['tags'].apply(lambda x:" ".join(x))
```

```python
new_df.head()
```

```python
# lowercasing the all the tags columns
new_df['tags'] = new_df['tags'].apply(lambda x:x.lower())
```

```python
#applying stemming on all the words in tags
import nltk
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
  y = []
  for i in text.split():
    y.append(ps.stem(i))

  return " ".join(y)
```

```python
new_df['tags']=new_df['tags'].apply(stem)
```

```python
# we have created the tags
# we will vectorize the tags using BOW
# we will remove all the stop words
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english')
```

```python
vectors = cv.fit_transform(new_df['tags']).toarray()
```

```python
vectors
```

```python
for i in cv.get_feature_names_out():
  print(i)
```

```python
# Importing Cosine similarity
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors)
```

```python
sorted(list(enumerate(similarity[0])),reverse=True,key = lambda x:x[1])
```

```python
 # creating a function such that when a movie name is given it provides 5 recommendation based on the similarity
def recommend(movie):
  movie_index = new_df[new_df['title']==movie].index[0]
  distances = similarity[movie_index]
  movies_list = sorted(list(enumerate(distances)),reverse=True,key = lambda x:x[1])[1:6]

  for i in movies_list:
    print(new_df.iloc[i[0]].title)
```

```python
# Our Recommend function is able to provide the 5 recommendaiton for the movies
recommend('Autumn in New York')
```
